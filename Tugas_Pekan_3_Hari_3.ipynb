```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, axs = plt.subplots(1, 3, sharex=True, sharey=True, figsize=(21,7))

l1 = axs[0].plot(x, tan, color="blue")[0]
l2 = axs[1].plot(x, cos, color="red")[0]
l3 = axs[2].plot(x, sin, color="green")[0]

axs[0].set_title('Grafik tanx')
axs[1].set_title('Grafik sinx')
axs[2].set_title('Grafik cosx')


fig.legend([l1, l2, l3],  ("tanX", "sinX", "cosX"), loc="upper center", borderaxespad=0.1)

plt.subplots_adjust(right=0.85)
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(16, 8))
ax.plot(x, y)

ins1 = ax.inset_axes([-0.1, 0.8, 0.3, 0.4])
ins1.plot(x, y, color='blue')
ins1.plot(x, y5, 'y')

ins2 = ax.inset_axes([-0.1, -0.2, 0.3, 0.4])
ins2.plot(x, y, color='blue')
ins2.plot(x, y2, color='red')

ins3 = ax.inset_axes([0.9, -0.2, 0.3, 0.4])
ins3.plot(x, y, color='blue')
ins3.plot(x, y3, color='green')

ins4 = ax.inset_axes([0.9, 0.8, 0.3, 0.4])
ins4.plot(x, y, color='blue')
ins4.plot(x, y4, color='black')

ax.set_title('Grafik CosX')
ins1.set_title('Grafik CosX^5')
ins2.set_title('Grafik CosX^2')
ins3.set_title('Grafik CosX^3')
ins4.set_title('Grafik CosX^4')

plt.show()
```
