Diketahu df3 merupakan DataFrame hasil gabung antara 'df1' dan 'df2' seperti pada gambar. Manakah perintah yang digunakan untuk menghasilkan 'df3'?

image.png

```
data1 = {
    'fruit' : ['Apple','Pear','Durian'],
    'price' : [4, 7, 10],
    'quantity' : [10, 14, 2]
}

data2 = {
    'fruit' : ['Apple','Durian','Grape'],
    'weight' : [10, 20, 2]
}

df1= pd.DataFrame(data1)
df2= pd.DataFrame(data2)
df3 = pd.merge(df1, df2, on='fruit', how='outer')
df3
```
