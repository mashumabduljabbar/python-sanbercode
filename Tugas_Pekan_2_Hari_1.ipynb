## Update Corona

```python
import requests
import pandas as pd
 
def getJSONData(apiURL):
   response = requests.get(
   apiURL,
   verify = True)
  
   response.raise_for_status()
 
   return(response.json())
  
apiURL = "https://api.kawalcorona.com/"
datakorona = getJSONData(apiURL)

b = []
for i in range(len(datakorona)):
  b.append(datakorona[i]['attributes'])

df = pd.DataFrame.from_dict(b)

df
```

```python
indo =df.loc[df['Country_Region'].values == 'Indonesia',:]
indo
```

```python
from datetime import datetime
a = indo['Last_Update'].values[0]/1000
print(f'Last update: {datetime.utcfromtimestamp(a)}')
```
