````python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
````

````python
df = pd.read_csv('https://raw.githubusercontent.com/multimediary/dataset/master/diamonds-1.csv')
````

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

````python
fig, ax = plt.subplots(figsize=(13, 13))
palette="ch:r=-.2,d=.3_r"

#code here
sns.set(style="ticks", color_codes=True)
ax = sns.scatterplot(df['carat'],df['price'], size=df['depth'], palette=palette,
                     hue_order=['I1','SI2','SI1','VS1','VS2','VVS1','VVS2','IF'],
                     data=df, sizes=(1,10), edgecolor = 'none', hue=df['clarity'])
````

# Soal 2: Facet Grid

Gunakan fungsi hexbin berikut untuk diterapkan pada mapping dari facetgridnya sehingga dihasilkan output sesuai expected.

```
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)
```


````python
#code here
sns.set(style="ticks")
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)
with sns.axes_style("dark"):
  g = sns.FacetGrid(df, hue="cut", col="cut", height=4)
g = g.map(hexbin, "carat", "price", extent=[0, 3.0 , 0, 20000])
````

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price sehingga dihasilkan output sesuai expected.

````python
df = df.sample(n=300, random_state=123)

#code here
g = sns.pairplot(df, hue="cut", vars=['carat','depth', 'table','price'], markers=["+","o", "s", "D","x"], corner=True)
````

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

````python
#code here
from scipy import stats

g = sns.jointplot("carat", "price", data=df,
                  kind="kde", space=0, color="b")


h = sns.jointplot("carat", "price", data=df, kind="reg", stat_func=stats.pearsonr)
````
